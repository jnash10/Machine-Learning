```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data.csv')
```

```python
data
```

```python
plt.scatter(data.iloc[:,0], data.iloc[:,1])
```

## DBSCAN

```python
from sklearn.cluster import DBSCAN
import numpy as np
```

```python
clusters = DBSCAN(eps=0.12, min_samples=30).fit(data)
plt.scatter(data.iloc[:,0], data.iloc[:,1], c=clusters.labels_)
```

```python
set(clusters.labels_)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans_clusters = KMeans(n_clusters=2, n_init=20, random_state=23).fit(data)
```

```python
plt.scatter(data.iloc[:,0], data.iloc[:,1], c=kmeans_clusters.labels_)
```

```python
training_instances = pd.read_csv('/home/agam/classes/ml/assignment1_data/training_data.csv')
training_labels = pd.read_csv('/home/agam/classes/ml/assignment1_data/training_data_class_labels.csv')

data2 = pd.DataFrame()
data2['x']=training_instances.iloc[:,0]
data2['y']=training_instances.iloc[:,1]
data2['labels']=training_labels
```

```python
plt.scatter(data2.iloc[:,0], data2.iloc[:,1],c=data2['labels'])
```

```python
X = data2.drop(['labels'],axis=1)
```

```python
clusters = DBSCAN(eps=0.1515, min_samples=60).fit(X)
plt.scatter(data2.iloc[:,0], data2.iloc[:,1], c=clusters.labels_)
```

```python
from sklearn.metrics.cluster import normalized_mutual_info_score
```

```python
normalized_mutual_info_score(data2['labels'], clusters.labels_)
```

```python
clusters = KMeans(n_clusters=2, n_init=20, random_state=23).fit(X)
plt.scatter(data2.iloc[:,0], data2.iloc[:,1], c=clusters.labels_)
```

```python
normalized_mutual_info_score(data2['labels'], clusters.labels_)
```
